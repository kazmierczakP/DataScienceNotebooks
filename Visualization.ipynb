### Diverging bar

### Surplus/Deficit filled line

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(seed=5)
index=pd.date_range(start='2020-01-01',end='2020-02-01')
df=pd.DataFrame(np.random.randn(32,1),index=index,columns=['col1'])
df=df.reset_index()
columns=list(df.columns)
columns[0]='date'
df.columns=columns
df['zero']=0
df.head()
```

```python
fig = plt.figure(figsize=(14, 10))
plt.plot(df['date'],df['col1'],df['date'],df['zero'],color='blue',linewidth=0.1)
plt.fill_between(df['date'],df['col1'],df['zero'],where=df['zero']<df['col1'],facecolor='cyan',interpolate=True)
plt.fill_between(df['date'],df['col1'],df['zero'],where=df['zero']>df['col1'],facecolor='blue',interpolate=True)
plt.show()
```

### Connected Scatterplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(seed=5)
df=pd.DataFrame(data=np.random.randn(6,2),columns=['x','y'])
u = np.diff(df['x'])
v = np.diff(df['y'])
midpoints_x=df['x'][:-1]+u/2
midpoints_y=df['y'][:-1]+v/2
plt.plot(df['x'],df['y'],marker='o',color='cyan')
plt.quiver(midpoints_x,midpoints_y,u,v, pivot="mid", color="cyan")
```

### Heatmap

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = [ "Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])
```

```python
plt.imshow(harvest)
plt.xlabel=farmers
plt.ylabel=vegetables
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = plt.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")
```

```python
import seaborn as sns
```

```python
df=pd.DataFrame(harvest,index=vegetables,columns=farmers)
```

```python
sns.heatmap(df, annot=True)
```

### Population pyramid

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(seed=1)
df=pd.DataFrame(np.random.random_sample((10,2)),columns=['male','female']).reset_index()
df=df.cumsum()
df['male']*=-1
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.hlines(y=range(10),xmin=0,xmax=df['male'][::-1],color='blue',linewidth=25)
plt.hlines(y=range(10),xmin=0,xmax=df['female'][::-1],color='cyan',linewidth=25)
```

### Cumulative curve

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(seed=1)
df=pd.DataFrame(np.random.random_sample((100,1)),columns=['x']).reset_index()
df['x']=df['x']*df.index*df.index
df=df.cumsum()
df.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['x'], linestyle='-')
```
