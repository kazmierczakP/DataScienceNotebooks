## timeseries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
index=pd.date_range('1/1/2000',periods=1000)
columns=['A','B','C','D']
```

```python
df=pd.DataFrame(np.random.randn(1000,4), index=index,columns=columns)
```

```python
df=df.cumsum()
```

```python
df.plot()
```

## facetting

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10)
fig=plt.figure()

fig.add_subplot(221)
plt.plot(np.cos(x))
fig.add_subplot(222)
plt.plot(np.sin(x))
fig.add_subplot(223)
plt.plot(np.tan(x))
fig.add_subplot(224)
plt.plot(x)
plt.show()
```

## grouping

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day = pd.read_csv('.\data\day.csv')
```

```python
plt.scatter('temp','cnt',data=day)
```

```python
plt.scatter('temp','cnt',data=day,c='season')
```

```python
fig=plt.figure()
fig.add_subplot(221)
plt.scatter('temp', 'cnt', data=day[day['season']==1], marker='o', color='blue')
fig.add_subplot(222)
plt.scatter('temp', 'cnt', data=day[day['season']==2], marker='v', color='green')
fig.add_subplot(223)
plt.scatter('temp', 'cnt', data=day[day['season']==3], marker='s', color='orange')
fig.add_subplot(224)
plt.scatter('temp', 'cnt', data=day[day['season']==4], marker='^', color='brown')
plt.show()
```
